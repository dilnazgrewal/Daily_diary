```python
from gtts import gTTS
from IPython.display import Audio, display
import random
import string
from datetime import datetime
```

```python
text = input("Enter the prompt: ")
```

```python
def unique_filename():
    time_stamp = datetime.now().strftime("%Y%m%d_%H%M%S")
    rand_str = ''.join(random.choices(string.ascii_letters, k=5))
    return f"output_{time_stamp}_{rand_str}.mp3"

filename = unique_filename()
```

```python
# Convert text to speech and save
tts = gTTS(text, lang='en')
tts.save(filename)

# Display playable audio in notebook
display(Audio(filename))

print(f"✅ Audio saved and ready to play: {filename}")
```
